### In this problem set, we will use this file from the USGS Earthquakes Database. The dataset is similar to the one you use in Assignment 02. Use the file provided (usgs_earthquakes.csv) to recreate the following map. Use the mag column for magnitude. [10 points]

```python
# Load modules
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Show plots in the notebook
%matplotlib inline
```

```python
# Open a csv file and see it
earth = pd.read_csv("usgs_earthquakes.csv")
top50=earth.sort_values("mag", ascending=False).head(50)
top50
```

```python
lat=top50['latitude']
lon=top50['longitude']
mag=top50['mag']
```

```python
# Create and define the size of a figure object 
fig = plt.figure(figsize=(14, 7))

ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180))
ax.set_global()
ax.stock_img()

plt.scatter(lon,lat,c=mag,cmap='Reds',edgecolors='k',vmin=6.5,vmax=8.2,transform=ccrs.PlateCarree())
plt.colorbar(shrink=0.5,label='magnitude')
plt.title('Top 50 earthquakes at 2014',fontsize=18)
plt.show()
```
