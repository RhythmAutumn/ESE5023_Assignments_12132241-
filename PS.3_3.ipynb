### 3.1 [5 points] Plot a time series of a certain variable with monthly seasonal cycle removed.

```python
# Load modules
import numpy as np
import math
import pandas as pd
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Show plots in the notebook
%matplotlib inline
```

```python
# Open a netCDF4 file and see it
ds = xr.open_dataset("temp_ITPCAS-CMFD_V0106_B-01_01dy_010deg_201601-201612.nc")
ds
```

```python
#Read variables
time=ds['time'][:]
lat=ds['lat'][:]
lon=ds['lon'][:]
temp=ds['temp'][:,:,:]
```

```python
#Select Wuhan and plot the time series
temp.sel(lon='114.23',method='nearest').sel(lat='30.40',method='nearest').plot()
plt.title('Daily near surface air temperature in Wuhan in 2016',fontsize=16)
plt.show()
```

### 3.2 [5 points] Make at least 5 different plots using the dataset.

```python
#Plot the average temperature versus latitude in 2016
temp.mean(axis=(0,2)).plot()
plt.title('The average temperature variation at different latitude in China in 2016')
plt.show()
```

```python
#Plot the average temperature versus latitude in eastern China in 2016 
temp.sel(lon=np.arange(105,140),method='nearest').mean(axis=(0,2)).plot()
plt.title('The average temperature variation at different latitude in eastern China in 2016')
plt.show()
```

### Comparing this graph to last graph, we find that the temperature decreases as the latitude becomes higher except the region of western China.

```python
#2D plot of the average near surface air temperature in China in 2016
temp.mean(axis=0).plot.contourf()
plt.title('The average near surface air temperature in China in 2016')
plt.show()
```

```python
#2D plot of the maximum near surface air temperature in China in 2016
temp.max(axis=0).plot.contourf()
plt.title('The maximum near surface air temperature in China in 2016')
plt.show()
```

```python
#2D plot of the minimum near surface air temperature in China in 2016
temp.min(axis=0).plot.contourf()
plt.title('The minimum near surface air temperature in China in 2016')
plt.show()
```
