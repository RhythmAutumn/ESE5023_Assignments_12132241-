#### 2.1 [5 points] Make a 2D plot of the time-mean TOA longwave, shortwave, and solar radiation for all-sky conditions. Add up the three variables above and verify (visually) that they are equivalent to the TOA net flux.

```python
# Load modules
import numpy as np
import math
import pandas as pd
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Show plots in the notebook
%matplotlib inline
```

```python
# Open a netCDF4 file and see it
ds = xr.open_dataset("CERES_EBAF-TOA_200003-201701.nc", engine="netcdf4")
ds
```

```python
#Read variables
time=ds['time'][:]
lat=ds['lat'][:]
lon=ds['lon'][:]
sw1=ds['toa_sw_all_mon'][:,:,:]
lw1=ds['toa_lw_all_mon'][:,:,:]
so1=ds['solar_mon'][:,:,:]
net1=ds['toa_net_all_mon'][:,:,:]
```

```python
#Calculate the mean and draw the 2D plot
sw1_mean=np.mean(sw1,axis=0)
lw1_mean=np.mean(lw1,axis=0)
so1_mean=np.mean(so1,axis=0)
net1_mean=np.mean(net1,axis=0)
new_mean=so1_mean-sw1_mean-lw1_mean

#Define the fields
fields = {0:'TOA SW for all sky conditions', 1:'TOA LW for all sky conditions', 2:'TOA solar radiation for all sky conditions', 3:'TOA net flux',4:'Addition of SW,LW and solar radiation'}

# Set the subgrids
fig = plt.figure(figsize=(16,8), dpi=120)
grid = plt.GridSpec(10, 20)  

# Plot each axes
plt.subplot(grid[0:4,0:6]) 
sw1_mean.plot()  
plt.title(fields[0])

# Plot each axes
plt.subplot(grid[0:4,7:13])  
lw1_mean.plot()    
plt.title(fields[1])

# Plot each axes
plt.subplot(grid[0:4,14:20]) 
so1_mean.plot()   
plt.title(fields[2])

# Plot each axes
plt.subplot(grid[5:10,0:9]) 
net1_mean.plot()   
plt.title(fields[3])

# Plot each axes
plt.subplot(grid[5:10,11:20]) 
new_mean.plot()   
plt.title(fields[4])

# Plot super title
plt.suptitle('2D plot of TOA SW,LW,solar radiation and net flux', verticalalignment='bottom', fontsize=20)   

# Show plot
plt.show()
```

### There are five graphs above. Graph 1-3 are 2D plots describing TOA shortwave,longwave and solar radiation for all sky conditions. Graph 4 is describing the TOA net flux. Graph 5 is describing incoming solar radiation minus the sum of TOA shortwave and longwave. Graph 4 and Graph 5 are the same, which indicates that the addition of the three variables above is equivalent to the TOA net flux.





### 2.2 [10 points] Calculate and verify that the TOA incoming solar, outgoing longwave, and outgoing shortwave approximately match up with the cartoon above.[Hint: Consider calculating the area of each grid]

```python
#import radius of earth and calculate the area of each grid
R=6371 #km
delta_phi=2*(math.pi)*R*(np.cos(lat*(math.pi)/180))/360
area=delta_phi*2*(math.pi)*R/360
area
# For grids in different latitudes, the area is different
```

```python
#Calculate the total surface area of the earth
Total_surface_area=area.sum()*360
Total_surface_area
```

```python
#Calculate the TOA outgoing shortwave
sw_lat=sw1_mean*area*1e6
outgoing_sw=sw_lat.sum()/Total_surface_area/1e6
outgoing_sw
```

### The TOA outgoing shortwave calculated is 99.139W/m^2, while the TOA outgoing shortwave in the cartoon is 77.0+22.9=99.9W/m**2. They can approximately match.

```python
#Calculate the TOA outgoing longwave
lw_lat=lw1_mean*area*1e6
outgoing_lw=lw_lat.sum()/Total_surface_area/1e6
outgoing_lw
```

### The TOA outgoing longwave calculated is 240.268W/m^2, while the TOA outgoing longwave in the cartoon is 239.9W/m**2. They can approximately match.

```python
#Calculate the TOA incoming solar
solar_lat=so1_mean*area*1e6
incoming_solar=solar_lat.sum()/Total_surface_area/1e6
incoming_solar
```

### The TOA incoming solar calculated is 340.285W/m^2, while the TOA incoming solar in the cartoon is 340.4W/m**2. They can approximately match.

### 2.3 [5 points] Calculate and plot the total amount of net radiation in each 1-degree latitude band. Label with correct units.

```python
#Calculate the net radiation
net_ra=so1_mean-lw1_mean-sw1_mean
net_ra#unit:W/m^2
```

```python
#Calculate the total amount of net radiation in each 1-degree latitude band
net_ra_band=net_ra*area*1e6
Total_net_ra_band=np.sum(net_ra_band,axis=1)#unit:W
Total_net_ra_band
```

```python
#Plot the total amount of net radiation in each 1-degree latitude band
fig = plt.figure(figsize=(10,5), dpi=70)
Total_net_ra_band.plot()
plt.title("The total amount of net radiation in each 1-degree latitude band",fontsize=12)
plt.ylabel('Net radiation (W)')
plt.show()
```
